```python
mpg = mpg.rename(columns = {'displ': 'displacement',
                      'cyl': 'cylinders',
                      'drv': 'drive'})
mpg
```

```python
mpg = mpg.drop(columns= ['displacement', 'cylinders'])
mpg
```

```python
mpg = mpg.assign(displacement = mpg.hwy - mpg.cty)
mpg
```

```python
plt.plot(mpg.cty)
```

```python
plt.scatter(mpg.cty, mpg.hwy)
plt.xlabel('city mpg')
plt.ylabel('highway mpg')
```

```python
plt.hist(mpg.hwy)
```

```python
g = mpg.groupby('drive').hwy.mean()
g
```
