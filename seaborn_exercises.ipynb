Load the chipotle dataset from SQL, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
from sqlalchemy import create_engine, text
from env import get_connection
```

```python
def get_db_url(database):
    return get_connection(database)
```

```python
url = get_db_url('chipotle')
query = text('SELECT * FROM orders')
engine = create_engine(url)
orders = pd.read_sql(query, engine.connect())
orders.head()
```

```python
orders['item_price'] = orders.item_price.str.replace('$', '').astype('float')
```

```python
cb = orders[orders.item_name == 'Chicken Bowl']
cb.sort_values('quantity', ascending = False)
```

```python
best = orders.groupby('item_name').quantity.sum().nlargest(4)
best
```

```python
revenue = orders[orders.item_name.isin(best.index)].groupby('item_name').item_price.sum()
revenue
```

```python
revenue = revenue.sort_values(ascending = False).reset_index()
revenue
```

```python
plt.figure(figsize = (8, 5))
sns.barplot(y = 'item_name', x = 'item_price', data = revenue)
plt.title('Best Selling Items')

plt.ylabel("") # to get remove the column_name
plt.xlabel("Total Revenue")
```
