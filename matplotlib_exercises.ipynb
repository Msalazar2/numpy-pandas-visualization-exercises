2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
    * y = √ x
    * y = x**3
    * y = 2 **x
    * y = 1 / ( x + 1 )

```python
x = np.linspace(10, 20, 100)


y1 = np.sqrt(x)
plt.plot(x, y1, label='y = √x')


y2 = x**3
plt.plot(x, y2, label='y = x^3')


y3 = 2 * x
plt.plot(x, y3, label='y = 2x')


y4 = 1 / (x + 1)
plt.plot(x, y4, label='y = 1/(x + 1)')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

4. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize = (14, 7))

x = np.linspace(0, 10, 100)


y1 = np.sqrt(x)

y2 = x**3

y3 = 2 * x

y4 = 1 / (x + 1)


plt.subplot(1, 4, 1)
plt.plot(x, y1)
plt.title('y1')

#define second subplot
plt.subplot(1, 4, 2)
plt.plot(x, y2)
plt.title('y2')

plt.subplot(1, 4, 3)
plt.plot(x, y3)
plt.title('y3')

plt.subplot(1, 4, 4)
plt.plot(x, y4)
plt.title('y4')

plt.show()
```

5. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python


# Define the range of x values
x = np.linspace(0, 10, 100)

# Equation 1: y = √x
y1 = np.sqrt(x)

# Equation 2: y = x^3
y2 = x**3

# Equation 3: y = 2x
y3 = 2 * x

# Equation 4: y = 1/(x + 1)
y4 = 1 / (x + 1)

# Create a figure with a single subplot
plt.figure(figsize=(10, 6))

# Plot each equation with a different color
plt.plot(x, y1, label='y1 = √x', color='blue')
plt.plot(x, y2, label='y2 = x^3', color='green')
plt.plot(x, y3, label='y3 = 2x', color='red')
plt.plot(x, y4, label='y4 = 1/(x + 1)', color='purple')

# Add labels to the axes
plt.xlabel('x')
plt.ylabel('y')

# Add a title to the figure
plt.title('Four Equations in One Figure')

# Add a legend to differentiate the equations
plt.legend()

# Show the combined figure
plt.grid(True)
plt.show()
```
