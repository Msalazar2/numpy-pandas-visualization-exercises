```python
import numpy as np
import pandas as pd
import math as math
from pydataset import data

import matplotlib.pyplot as plt
```

```python
x= (range(1,20))
y1 = [1 for n in x]
y2 = [math.log(n) for n in x]
y3 = x
y4= [(n * math.log(n)) for n in x]
y5= [n**2 for n in x]
y6= [2**n for n in x]
y7= [math.factorial(n) for n in x]
y8= [n**n for n in x]

plt.figure(figsize = (18,8))

plt.plot(x, y1, label='O(1)', color='blue')
plt.plot(x, y2, label='O(log n)', color='red')
plt.plot(x, y3, label='O(n)', color='purple')
plt.plot(x, y4, label='O(n log n)', color='green')
plt.plot(x, y5, label='O(n^2)', color='yellow')
plt.plot(x, y6, label='O(2^n)', color='black')
plt.plot(x, y7, label='O(n!)', color='orange')
plt.plot(x, y8, label='O(n^n)', color='silver')

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.annotate('LaTex', xy= (1.5,10))
plt.xlim(1, 11)
plt.ylim(0, 50)
```
